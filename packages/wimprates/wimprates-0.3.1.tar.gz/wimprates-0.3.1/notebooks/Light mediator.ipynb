```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import wimprates as wr
```

```python
es = np.linspace(0.01, 30, 100)
plt.plot(es, wr.rate_wimp_std(es, mw=100, sigma_nucleon=1e-46),
         label='$m_\phi = \infty$, $\sigma = 10^{-46} \mathrm{cm}^2$')
plt.plot(es, wr.rate_wimp_std(es, mw=100, m_med=0.01, sigma_nucleon=6.5e-44),
         label=r'$m_\phi = 10 \mathrm{MeV}/c^2$, $\sigma = 6.5 \times 10^{-44} \mathrm{cm}^2$',
         color='g')
plt.plot(es, wr.rate_wimp_std(es, mw=100, m_med=0.001, sigma_nucleon=6e-40),
         label=r'$m_\phi = 1 \mathrm{MeV}/c^2$, $\sigma = 6 \times 10^{-40} \mathrm{cm}^2$',
         color='r')

plt.legend(loc='upper right')
plt.xlabel('Recoil energy (keV)')
plt.ylabel('Events / (kg keV day)')
plt.yscale('log')
plt.ylim(1e-1, 1e2)
plt.title('$m_\chi = 200 \mathrm{GeV}/c^2$')
plt.savefig('light_mediator_effect.png')
```
